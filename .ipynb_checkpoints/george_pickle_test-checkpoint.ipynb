```python
import numpy as np
import pylab as plt
import george
import cPickle as pickle
%matplotlib inline
```

```python
# generate some data
N = 20
np.random.seed(1)
x = np.sort(np.random.rand(N)*100)
y = x + 50*np.sin(2*np.pi*x/30)
s = 25
y += np.random.randn(N)*s
yerr = np.repeat(s, N)

plt.errorbar(x, y, yerr, fmt='k.')
```

```python
# setup GP with optimized QP hyperparams
a, l, G, P = 100, 100, 1, 30
k1 = george.kernels.ExpSquaredKernel(l)
k2 = george.kernels.ExpSine2Kernel(G,P)
gp = george.GP(a*a*k1*k2)
gp.compute(x, yerr)
```

```python
# make prediction
x_pred = np.linspace(x.min(), x.max(), 500)
mu,cov = gp.predict(y, x_pred)

plt.errorbar(x, y, yerr, fmt='k.')
plt.plot(x_pred, mu, 'b-')
```

```python
# pickle object
fname = 'tmp_george'
do_I_pickle = True
if do_I_pickle:
    f = open(fname, 'wb')
    pickle.dump(gp, f)
    f.close()
```

```python
# load pickle
if do_I_pickle:
    f = open(fname, 'rb')
    gp = pickle.load(f)
    f.close()
```

```python
# remake prediction
x_pred = np.linspace(x.min(), x.max(), 500)
mu,cov = gp.predict(y, x_pred)

plt.errorbar(x, y, yerr, fmt='k.')
plt.plot(x_pred, mu, 'r-')
```
